```python
import pandas as pd

df = pd.read_excel('Analisis de riesgo Crediticio- DATA SCIENCE I.xlsx')
```

**Abstract**
El dataset analizado contiene información detallada sobre clientes y sus cuentas de préstamo, incluyendo variables como el score crediticio (clasificado como "Bajo", "Regular" o "Bueno"), días de mora, monto del préstamo, tipo de producto crediticio, canales de solicitud y desembolso, y datos demográficos como género, fecha de nacimiento, provincia y ciudad de residencia. Este trabajo tiene como objetivo principal anticipar el comportamiento de pago de los clientes mediante el análisis de la relación entre el monto del préstamo y los días de mora, con el fin de mejorar la gestión del riesgo de impago. Además, se busca identificar los canales de solicitud y desembolso más eficientes para optimizar la estrategia crediticia y reducir la morosidad.

Otro objetivo clave es segmentar a los clientes según su ubicación geográfica, productos solicitados y características demográficas, lo que permitirá desarrollar ofertas más personalizadas y mejorar la fidelización. Asimismo, se pretende detectar tendencias de consumo para diseñar estrategias comerciales alineadas con la demanda del mercado. Finalmente, se evaluará la relación entre el monto del préstamo y los días de mora, con el objetivo de ajustar las políticas crediticias y mejorar la gestión del riesgo financiero.

Las **hipótesis planteadas** sugieren que:

H1: Los clientes con préstamos de montos más altos tienden a tener mayores días de mora.

H2: Los canales de solicitud y desembolso más utilizados están asociados a menores tasas de morosidad.

H3: La ubicación geográfica de los clientes influye en el comportamiento de pago, con ciertas regiones mostrando mayores tasas de morosidad.

H4: Los clientes de género femenino tienen un comportamiento de pago más responsable (menos días de mora) que los clientes de género masculino

Grafico de visualizacion para la pregunta: ¿Existe una relación entre el monto del préstamo y los días de mora?

```python
import seaborn as sns
import matplotlib.pyplot as plt

#H1:
# Scatterplot
sns.scatterplot(x='MONTO', y='DIAS_MORA', data=df)
plt.title('Relación entre monto del préstamo y días de mora')
plt.xlabel('Monto del préstamo')
plt.ylabel('Días de mora')
plt.show()
```

El scatterplot muestra la relación entre el monto del préstamo y los días de mora. A simple vista, no se observa una tendencia clara: los días de mora parecen distribuirse de manera similar para todos los montos de préstamo. Esto sugiere que el monto del préstamo no es un factor determinante en la morosidad. Sin embargo, para confirmar esta hipótesis, es necesario calcular la correlación entre ambas variables.

Grafico de visualizacion de la pregunta: ¿Qué canales de solicitud y desembolso están asociados a menores tasas de morosidad?

```python
#H2
# Gráfico de barras
sns.barplot(x='ID_CANAL_SOLICITUD', y='DIAS_MORA', data=df, estimator=sum)
plt.title('Días de mora por canal de solicitud')
plt.xlabel('Canal de solicitud')
plt.ylabel('Días de mora totales')
plt.xticks(rotation=45)
plt.show()
```

El gráfico de barras muestra los días de mora totales por canal de solicitud. Se observa que el canal App tiene la menor cantidad de días de mora, lo que sugiere que es el más eficiente. Por otro lado, el canal Web tiene la mayor cantidad de días de mora, lo que podría indicar problemas en el proceso de solicitud o desembolso. Esto respalda la hipótesis de que ciertos canales están asociados a menores tasas de morosidad.

Gráfico de visualizacion ¿Cómo varía el comportamiento de pago de los clientes según su ubicación geográfica?

```python
#H3
# Agrupar datos por Provincia
mora_por_region = df.groupby('ID_PROVINCIA')['DIAS_MORA'].mean().reset_index()

# Gráfico de barras
sns.barplot(x='ID_PROVINCIA', y='DIAS_MORA', data=mora_por_region)
plt.title('Días de mora promedio por provincia')
plt.xlabel('Provincia')
plt.ylabel('Días de mora promedio')
plt.xticks(rotation=90)
plt.show()
```

El gráfico de barras muestra el promedio de días de mora por provincia. La provincia San Luis tiene el mayor promedio de días de mora, lo que podría estar relacionado con factores socioeconómicos o de acceso a servicios financieros. En contraste, la provincia de Formosa tiene el menor promedio, lo que sugiere un mejor comportamiento de pago en esa región. Esto respalda la hipótesis de que la ubicación geográfica influye en el comportamiento de pago.
